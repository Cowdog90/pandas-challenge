```python
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
sch_data = "Resources/schools_complete.csv"
stu_data = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(sch_data)
student_data = pd.read_csv(stu_data)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
complete_data = pd.DataFrame(school_data_complete)

#complete_data['% Passing Math']=complete_data[('math_score')>=70.0].count()


complete_data.rename(columns = {'Student ID':'Student ID', 'student_name':'Student Name','gender':'Gender','grade':'Grade','school_name':'School Name','reading_score':'Reading Score','math_score':'Math Score','School ID':'School ID','type':'Type','size':'Size','budget':'Budget'}, inplace = True) 
   
# Count the number of schools
unique_value = school_data_complete["school_name"].nunique() 

# sum the number of students and sum the total budget

student_num = school_data_complete['student_name'].count()
 
total_budget = school_data['budget'].sum()

# find avereage math and reading scores

ave_math = school_data_complete['math_score'].mean()

ave_read = school_data_complete['reading_score'].mean()

# find percentage of passing math scores

math_df = student_data.loc[student_data['math_score'] >= 70]

math_cnt = math_df['math_score'].count()

math_per = (math_cnt/student_num)*100

# find percentage of passing reading scores

read_df = student_data.loc[student_data['reading_score'] >= 70]

read_cnt = read_df['math_score'].count()

read_per = (read_cnt/student_num)*100

# calculate overall passing rate

overall_pass = ((read_per+math_per)/2)



# stuff it into a dataframe

# Count the number of schools
#unique_value = school_data_complete["school_name"].nunique() 

columns = ['Number of Schools', 'Total Students', 'Total Budget', 'Average Math Score', ' Average Reading Score', '% Passing Math', '% Passing Reading', '% Passing']
values = [unique_value, student_num, total_budget, ave_math, ave_read, math_per, read_per, overall_pass]
 
district_summary = pd.DataFrame(columns = ['Number of Schools', 'Total Students', 'Total Budget', 'Average Math Score', ' Average Reading Score', '% Passing Math', '% Passing Reading', '% Passing']) 
district_summary = district_summary.append({'Number of Schools': unique_value, 'Total Students': student_num, 'Total Budget': total_budget, 'Average Math Score' : ave_math, 'Average Reading Score' : ave_read, '% Passing Math': math_per, '% Passing Reading': read_per, '% Passing': overall_pass},values)
district_summary.head()
```

```python
per_passing_math=complete_data.groupby('School Name')['Math Score'].apply(lambda c: (c>=70).sum()/len(c))*100
per_passing_read=complete_data.groupby('School Name')['Reading Score'].apply(lambda c: (c>=70).sum()/len(c))*100

per_passing_both = pd.merge(per_passing_math, per_passing_read, how="left", on=["School Name"])

per_passing_both.rename(columns = {'Math Score':'% Passing Math', 'Reading Score':'% Passing Reading'}, inplace = True)

per_passing_both['% Total Passing']=per_passing_both.mean(axis=1)

```

```python
school_summary=complete_data.groupby('School Name').agg({'Type':'max','Size':'max','Budget':'max','Reading Score':'mean','Math Score':'mean'})

sch_sum = pd.merge(school_summary, per_passing_both, how="left", on=["School Name"])

sch_sum['Budget/Student']=(sch_sum['Budget'] / sch_sum['Size'] ) 
   
sch_sum.head(20)
```

```python
best_schools = sch_sum.sort_values('% Total Passing', ascending=[False])

best_schools.head(5)
```

```python
# Breakout dataframe by grade.

ninth = complete_data[complete_data.Grade == '9th']
tenth = complete_data[complete_data.Grade == '10th']
eleventh = complete_data[complete_data.Grade == '11th']
twelfth = complete_data[complete_data.Grade == '12th']
```

```python
#group new datasets by school and get the mean reading and math scores

ninth = ninth.groupby('School Name').agg({'Type':'max','Size':'max','Budget':'max','Reading Score':'mean','Math Score':'mean'})
tenth = tenth.groupby('School Name').agg({'Type':'max','Size':'max','Budget':'max','Reading Score':'mean','Math Score':'mean'})
eleventh = eleventh.groupby('School Name').agg({'Type':'max','Size':'max','Budget':'max','Reading Score':'mean','Math Score':'mean'})
twelfth = twelfth.groupby('School Name').agg({'Type':'max','Size':'max','Budget':'max','Reading Score':'mean','Math Score':'mean'})
```

```python
# drop unneeded columns

ninth = ninth.drop(['Type','Size','Budget'], axis = 1)
tenth = tenth.drop(['Type','Size','Budget'], axis = 1)
eleventh = eleventh.drop(['Type','Size','Budget'], axis = 1)
twelfth = twelfth.drop(['Type','Size','Budget'], axis = 1)

# rename remaining columns

ninth.rename(columns = {'Reading Score':'9th Grade Reading', 'Math Score':'9th Grade Math'}, inplace = True)
tenth.rename(columns = {'Reading Score':'10th Grade Reading', 'Math Score':'10th Grade Math'}, inplace = True)
eleventh.rename(columns = {'Reading Score':'11th Grade Reading', 'Math Score':'11th Grade Math'}, inplace = True)
twelfth.rename(columns = {'Reading Score':'12th Grade Reading', 'Math Score':'12th Grade Math'}, inplace = True)
```

```python
all_sch_byGrade = pd.concat([ninth, tenth, eleventh, twelfth], axis=1)

#all_sch_byGrade.head(15)
```

```python
# Spilt dataframe into a math and reading dataframe

mathByGrade = all_sch_byGrade.copy()
readByGrade = all_sch_byGrade.copy()

# Drop reading from math and drop math from reading

readByGrade = all_sch_byGrade.drop(['9th Grade Math','10th Grade Math','11th Grade Math','12th Grade Math'], axis = 1)
mathByGrade = all_sch_byGrade.drop(['9th Grade Reading','10th Grade Reading','11th Grade Reading','12th Grade Reading'], axis = 1)


```
